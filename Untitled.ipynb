```python
import pandas as pd
import numpy as np
import datetime
from bokeh.charts import output_notebook, show, output_file
from bokeh.plotting import Figure,ColumnDataSource
output_notebook()

df = pd.read_excel('veg.xls')
df['產品大項']=df['產品名稱'].str.split(' ').str.get(0)
df['產品小項']=df['產品名稱'].str.split(' ').str.get(1)
df['西元日期']=[str(int(x.split('/')[0])+1911)+'/'+(x.split('/')[1])+'/'+(x.split('/')[2]) for x in df['日期']]

byDate = df.groupby(['西元日期'],as_index=False).sum()
byMarket = df.groupby(['市場'],as_index=False).sum()
print(byMarket)
#byDate_Market.sort(['西元日期'], ascending=[True])


def get_data(dataFrame):
    return ColumnDataSource(data=dict(
            date = pd.to_datetime(dataFrame['西元日期']),
            market = [str(x) for x in dataFrame['市場']],
            price = [str(x) for x in dataFrame['交易金額(元)']],
            qty = [str(x) for x in dataFrame['交易量(公斤)']],
            #tipdate=[x.strftime("%Y-%m-%d") for x in df['日期']]
        ))

#print(get_data(byMarket).data['market'])
f = Figure(plot_width=800, plot_height=300, x_axis_type="datetime", x_axis_label='日期', y_axis_label='價格')
f.line('date', 'qty', line_width=3, line_color='red', source = get_data(byDate))
f.line('date', 'qty', line_width=3, line_color='red', source = get_data(byMarket))

#f.vbar(get_data(byMarket).data['qty'],get_data(byMarket).data['market'], width=0.5, bottom=0, top=get_data(byMarket).data['qty'], color="#CAB2D6", source = get_data(byMarket))

show(f)
```
